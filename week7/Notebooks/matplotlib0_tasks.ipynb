```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. CREATE THE DATA (The part that was missing!)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [19, 22, 20, 23, 21],
    'Score': [85, 78, 92, 88, 76]
}
df = pd.DataFrame(data)

# 2. NOW PLOT IT
plt.figure(figsize=(8, 5))
plt.plot(df['Name'], df['Score'], marker='o', color='b', linestyle='-')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.title('Student Scores Trend')
plt.show()
```

```python
# Create the Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(df['Name'], df['Score'], color='skyblue', edgecolor='navy')

# Customization
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.title('Comparison of Student Scores')

# Display the plot
plt.show()
```
